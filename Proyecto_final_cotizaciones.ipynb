```python
import pandas as pd
import numpy as np
import requests
import krakenex
import plotly.graph_objects as go
from pykrakenapi import KrakenAPI
import ta
```

```python
api = krakenex.API()
k = KrakenAPI(api)

pairs = k.get_tradable_asset_pairs()
pairs.head().T
```

```python
def leer_moneda (pairs_altname):   
    moneda = input('Ingresar moneda: ')
    
    if moneda not in pairs_altname:
        
        raise ValueError ('Par seleccionado no se encuentra entre las cotizaciones de la tabla')

    return moneda
```

```python
def mostrar_grafico(moneda):
  ohlc = k.get_ohlc_data(moneda, interval=1440, since=1628121600, ascending = True)
  ohlc[0]['Mean'] = ohlc[0]['close'].rolling(20).mean() #es la media
  ohlc[0]["rsi"] = ta.momentum.rsi(ohlc[0]["close"], window=14, fillna=False) #importacion de libreria TA
  layoutt = go.Layout(autosize=False, width=4181, height=1597)
# make it fit on my screen!!!
  layoutt = go.Layout(autosize=True)
  layoutt2 = go.Layout(autosize=False, width=6731, height=2571)
  fig = go.Figure(data=[go.Candlestick(x = ohlc[0].index,    #las velas crecientes en verde y las decrecientes en rojo
                                    open = ohlc[0]['open'],
                                    high = ohlc[0]['high'],
                                    low = ohlc[0]['low'],
                                    close = ohlc[0]['close'],
                                    ),
                        go.Scatter(x=ohlc[0].index, y=ohlc[0]["rsi"], mode="markers+lines", name="RSI", yaxis="y2"), #ayuda a generar otro eje
                        go.Scatter(x=ohlc[0].index, y=ohlc[0]['Mean'], line=dict(color='black', width=1))],
                 layout=layoutt).update_layout(
    yaxis_domain=[0.3, 1],
    yaxis2={"domain": [0, 0.20]},
    xaxis_rangeslider_visible=False,
    showlegend=False,
)
                                              
  fig.show()
```

```python
try: 
  moneda = leer_moneda(pairs['altname'])

  mostrar_grafico(moneda)   

except ValueError as e:
  print(e)
```
